#### Find all the information that was updated on 15 of July 2021

```python
july15 = covid_vaccine[covid_vaccine.Date == '15/07/2021']
july15
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.barplot(y='State',x='Total Doses Administered',data = july15,palette="hot")
plt.title("Barplot of states and doses administered", fontsize=20)
plt.xlabel("Total Doses", fontsize=15)
plt.ylabel("State", fontsize=15)
plt.show()
```
